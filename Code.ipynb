```python
import cv2
from matplotlib import pyplot as plt

# Load the pre-trained face detection classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the CV image
cv_image = cv2.imread(r'Te.jpeg')

# Convert BGR image to RGB for displaying in Jupyter
cv_rgb = cv2.cvtColor(cv_image, cv2.COLOR_BGR2RGB)

# Display the CV image
plt.imshow(cv_rgb)
plt.axis('off')
plt.show()

# Convert the CV image to grayscale
cv_gray = cv2.cvtColor(cv_image, cv2.COLOR_BGR2GRAY)

# Detect faces in the CV image
faces = face_cascade.detectMultiScale(cv_gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Extract faces and display them
for i, (x, y, w, h) in enumerate(faces):
    face = cv_image[y:y+h, x:x+w]
    cv2.imwrite(f'face_{i}.jpg', face)
    
    # Convert BGR face image to RGB for displaying in Jupyter
    face_rgb = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
    
    # Display the extracted face
    plt.imshow(face_rgb)
    plt.axis('off')
    plt.title(f'Extracted Face {i+1}')
    plt.show()
```
